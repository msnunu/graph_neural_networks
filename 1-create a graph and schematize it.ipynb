```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#create an empty graph
G = nx.Graph()
```

```python
#add manually nodes to G graph
G.add_node("A")
G.add_node("B")
```

```python
#add nodes from list
G.add_nodes_from(["D", "E"])
```

```python
#add nodes by using loop
node_list= ["G", "H"]
for node in node_list:
    G.add_node(node)
```

```python
#adding edges
edges = [ ("A", "B"), ("A", "E"), ("A", "D"), ("B", "H"), ("E" "D"), ("D", "H"), ("G", "H")]
G.add_edges_from(edges)
```

```python
#just create node positions
pos = nx.spring_layout(G)
```

```python
#drawing graph
nx.draw_networkx(G, pos, with_labels=True, node_size=300, node_color="skyblue")
```

### creating directed graph with weighted edges and nodes 

```python
G2= nx.DiGraph()
```

```python
G2.add_node('A', weight=0.5)
G2.add_node('B', weight=0.8)
G2.add_node('C', weight=0.6)
```

```python
G2.add_edge('A', 'B', weight=0.3)
G2.add_edge('B', 'C', weight=0.7)
G2.add_edge('C', 'A', weight=0.9)
```

```python
nx.draw_networkx(G2, with_labels=True, node_size=500, node_color="lightblue")
labels = nx.get_edge_attributes(G2,'weight')
nx.draw_networkx_edge_labels(G2, pos=nx.spring_layout(G2), edge_labels=labels)
plt.show
```

```python
pos2 = nx.spring_layout(G2)
nx.draw_networkx(G2, pos2, with_labels=False, node_size=2000, node_color='lightblue', font_weight='bold')
labels = nx.get_edge_attributes(G2, 'weight')
nx.draw_networkx_edge_labels(G2, pos2, edge_labels=labels)
node_labels = nx.get_node_attributes(G2, 'weight')
nx.draw_networkx_labels(G2, pos2, labels=node_labels)
plt.show()
```
